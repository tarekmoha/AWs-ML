# What is machine learning.

- AI :Cababilty of machines to perform a human like activities.
- ML: Type of AI allows computer to learn and improve from experience.
- Supervised learning: Every Entry in the data has label.
- Unsupervised: No labels
- Reinforcement learning: Learning with actions.

### Components of ML:

- ML Model: Model can be sheped into too many shapes and serve too many problems, 
    - Generic program [ block of code ] made specific by data
    - i.e. Linear regression.
- Training Algorithm: make adjustments to make it the desired shape.
    - process to use a data to shape the model
- Model Inference algorithm: Evaluate it and use it.
    - Use of the model to solve a task
